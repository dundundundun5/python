SETUP

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('iris')
```

1. sns.replot(x=None,y=None,data=None, **kwargs)

    用于绘制两个变量之间的关系及子集的语义映射，默认绘制散点图
    

```python
# creating the replot
sns.relplot(x='sepal_width', y='species', data=data)
```

2. sns.scatterplot(x=None,y=None,data=None,**kwargs)

    散点图，使用点云描述两个变量的联合分布
    <https://www.geeksforgeeks.org/scatterplot-using-seaborn-in-python/>
    <https://www.geeksforgeeks.org/visualizing-relationship-between-variables-with-scatter-plots-in-seaborn/>

```python
# creating the scatterplot
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
```

3. sns.lineplot(x=None,y=None,data=None,**kwargs)

    折线图，多用于显示变量与变量之间的变化关系，比如指标-时间，误差-超参

```python
# creating lineplot
sns.lineplot(x='sepal_length', y='species', data=data)
```

4. sns.barplot(x,y,hue,data,order,hue_order)

    柱状图，一般用于根据某些方法聚合分类数据，为x选择分类列，为y选择数字列

```python
# creating barplot
sns.barplot(x='species', y='sepal_length', data=data)
```

5. countplot(x,y,hue,data,order,..)

    countplot统计类别并返回其出现次数。
    <https://www.geeksforgeeks.org/countplot-using-seaborn-in-python/>

```python
# creating countplot
sns.countplot(x='species', data=data)
```

6. boxpplot(x,y,hue,data,hue_order,orient)

    箱线图表示变量之间比较的定量数据分布，图像部分表示正常值的区间，实线部分表示异常值的点
    
    <https://www.geeksforgeeks.org/boxplot-using-seaborn-in-python/>

```python
# creating boxplot
plt.subplot(211)
sns.boxplot(x='species', y='sepal_width', data=data)
plt.subplot(212)
# creating horizontal boxplot
sns.boxplot(y='species', x='sepal_width', data=data)
```

7. sns.violoinplot(x,y,hue,data,order,...)

    小提琴图，是一种高级的箱线图。在数据的不同值出显示不同的概率密度
    <https://www.geeksforgeeks.org/violinplot-using-seaborn-in-python/>

```python
sns.violinplot(x='species', y='sepal_width',data=data)
```

8. sns.stripplot(x,y,hue,data,...)

    分类散点图，只是分类
