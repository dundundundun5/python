先来一个简单例子

```python
# importing pkg
import seaborn as sns
import matplotlib.pyplot as plt
# loading iris dataset
data = sns.load_dataset('iris')
# draw
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
```

给此图添加一个标题

```python

sns.lineplot(x='sepal_length', y='sepal_width', data=data)

# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
```

setting the x limit

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
plt.xlim(5) # show the graph with x >= 5 only
```

改变图的风格

    set_style(style=,rc=)

    style= 'darkgrid'| 'whitegrid'| 'dark'| 'white'| 'ticks'


```python
sns.lineplot(x='sepal_length', y='sepal_width', data=data)

sns.set_style('dark')
```

轴是表示数据边界并连接轴刻度线的线。可以使用despine（）方法将其删除。

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

sns.despine() # 移除了坐标轴
```

设置画布的大小

```python
# 设置画布大小
plt.figure(figsize=(2, 4))
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
sns.despine()
```

缩放绘图

    set_context(context=None, font_scale=1, rc=None)
    context='notebook', 'paper', 'talk', 'poster'
    font_scale: float 放大/缩小 倍数


```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
# 风格为 'poster' 缩放比例变为原来的2倍
sns.set_context('poster', font_scale=2)
```

1. 暂时修改坐标轴的样式

    axes_style(style=None, rc=None) 
    style = 同上

```python
def plot():
    sns.lineplot(x='sepal_length', y='sepal_width', data=data)

with sns.axes_style('darkgrid'):

    # Adding the subplot
    plt.subplot(211)
    plot()

plt.subplot(212)
plot()
```

3. palplot()

    可以将调色板的内容用绘图的方式画出来
4. set_palette()

    给所有图片的线条设置成调色板的颜色, 颜色可以是list[str]也可以是str

<>

```python
# 获取当前调色板
palette = sns.color_palette()

# 画出调色板的内容
sns.palplot(palette)
```

```python
# 5个颜色点，左边粉红色，右边绿色， 从两边向中间逐渐变淡
palette = sns.color_palette('PiYG', 5)

sns.palplot(palette)
```

```python
# 16个颜色块，从浅色到深色
palette = sns.color_palette('Greens', 16)

sns.palplot(palette)
```

```python
sns.set_palette('vlag')
plt.subplot(211)

# plotting with the color palette as vlag
plot()

sns.set_palette('Accent')
plt.subplot(212)
plot()
```
